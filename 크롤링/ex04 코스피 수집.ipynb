# 종목명, 현재가격, 거래량
## 데이터프레임 제작, csv저장

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
res = req.get("https://finance.naver.com/sise/sise_quant.naver")
```

```python
soup = bs(res.text,"lxml")
co = soup.select('a.tltle')
num = soup.select('td.number:nth-child(3)')
tv = soup.select('td.number:nth-child(6)')
```

```python

for i in co:
    print(i.text)
for i in num:
    print(i.text)
for i in tv:
    print(i.text)
```

```python
eventName = []
pricing = []
tradingVolume = []
for i in range(len(co)):
    eventName.append(co[i].text)
    pricing.append(num[i].text)
    tradingVolume.append(tv[i].text)
    
```

```python
import pandas as pd
```

```python
dic = {"종목명" : eventName, "가격정보" : pricing, "거래량":tradingVolume}
kospi = pd.DataFrame(dic)
kospi
```

```python
kospi.to_csv("코스피.csv", encoding="euc-kr")
```
