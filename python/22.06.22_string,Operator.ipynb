### 산술 연산자

```python
x = 7
y = 2

print(x + y) # 덧셈
print(x - y) # 뺄셈
print(x * y) # 곱셈
print(x / y) # 나눗셈
print(x % y) # 나머지
print(x // y) # 나누기 몫
```

```python
num1 = 10
num2 = 7

print(num1/num2)  # 나누기
print(num1%num2)  # 나머지
print(num1//num2) # 나누기(몫)
```
